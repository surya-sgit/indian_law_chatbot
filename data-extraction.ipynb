```python
import pytesseract
from pdf2image import convert_from_path
```

```python
pdf_path = "/kaggle/input/ipc-data/Indian Penal Code Book (2).pdf"
batch_size = 50  
total_pages = 1871

for start in range(1, total_pages + 1, batch_size):
    end = min(start + batch_size - 1, total_pages)
    print(f"Processing pages {start} to {end}...")

    pages = convert_from_path(pdf_path, dpi=150, first_page=start, last_page=end)  

    for i, page in enumerate(pages, start=start):
        page.save(f"page_{i}.jpg", "JPEG")

print("Conversion complete!")
```

```python
import os

text_data = ''

for i in range(1, 1872):  
    image_path = f"page_{i}.jpg"
    
    if os.path.exists(image_path):  
        text = pytesseract.image_to_string(image_path)  
        text_data += text + '\n'  

with open("extracted_text.txt", "w", encoding="utf-8") as f:
    f.write(text_data)

print("Text extraction complete")
```
